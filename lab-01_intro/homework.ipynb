```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [4, 4]
```

```python
img = cv2.imread('data/landscape-pixel-art.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis(False)
```

```python
red, green, blue = cv2.split(img)

img1 = cv2.merge([red, green, blue])
img2 = cv2.merge([red, blue, green])
img3 = cv2.merge([green, red, blue])
img4 = cv2.merge([blue, green, red])

out1 = np.hstack([img1, img2])
out2 = np.hstack([img3, img4])
out = np.vstack([out1, out2])

plt.imshow(out)
plt.axis(False)
```

Для виведення зображення на екран використовується формат RGB, тож використовуючи змінений порядок каналів, кольори змінюватимуться. На приклад, жовтий колір на оригінальному зображені (утворюється змішуванням червоного та зеленого) стане рожевим, якщо зелений та синій канали поміняти місцями. Використовуючи порядок каналів, що не передбачений скажімо згортковою моделлю, шукані ознаки можуть виявитися хибними. На приклад, модель навчена на даних "розуміє", що на зображенні трави переважає зелений канал, тож обробляючи зображення, де переважним буде синій риси трави будуть не помічені
